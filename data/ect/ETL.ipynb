```python
import pandas as pd
```

```python
df = pd.read_csv(r"ect.csv", engine='pyarrow')
df.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(df):
    # Remove leading and trailing whitespace in column: 'unitName'
    df['unitName'] = df['unitName'].str.strip()
    
    # to title case, cols
    # df.columns = df.columns.str.title()
    df.columns = map(lambda x: x[0].upper() + x[1:], df.columns)

    return df

df_clean = clean_data(df.copy())
df_clean.head()
```

```python
df_clean.to_parquet("ect.parquet", index=False)
```
