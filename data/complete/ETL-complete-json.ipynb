```python
# Custom function to filter groups with count == 1
def filter_count_1(group):
    return group.count().iloc[0] > 1

# Group by the specified columns and apply the custom filter function
result = df.groupby(['จังหวัด','เขตเลือกตั้ง','สำนักทะเบียน', 'ตำบล', 'หน่วยเลือกตั้ง']).filter(filter_count_1)

result
```

```python

def clean_data(df):
    # Split text using string '#' in column: 'สถานที่เลือกตั้ง'
    df['สถานที่เลือกตั้ง'] = df['สถานที่เลือกตั้ง'].str.replace('#', '')

    # Rename column 'สถานที่เลือกตั้ง_0' to 'สถานที่เลือกตั้ง'
    df = df.rename(columns={
        'รหัสจังหวัด': 'Province_Code',
        'จังหวัด': 'Full_Province',
        'เขตเลือกตั้ง': 'Division_Number',
        'รหัสสำนักทะเบียน': 'Registration_Office_Code',
        'สำนักทะเบียน': 'Registration_Office',
        'รหัสตำบล': 'Sub_District_Code',
        'ตำบล': 'Tambon',
        'สถานที่เลือกตั้ง': 'Street_Address',
        'หน่วยเลือกตั้ง': 'Unit_Number',
        })

    return df


df_clean = clean_data(df.copy())
df_clean.head()
```
